# All Respondents

**Ratings on a scale of 1 to 10**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import MaxNLocator
```

```python
df = pd.read_excel('KGNU-Member-Data-06.14.22.xlsx', sheet_name='English') 
```

```python
cols =  df.columns.tolist()
#cols
```

```python
show_ratings_cols = df.columns.tolist()[111:191]

for col in show_ratings_cols:
    df[col].value_counts().plot.bar(rot=90)
    plt.title(col)
    
    axes = plt.gca()       
    ya = axes.get_yaxis()
    ya.set_major_locator(MaxNLocator(integer=True))
    
    #plt.xlabel('Rating')
    plt.ylabel('Number of Respondents')
    
    plt.show()

    ra = df[col].apply(pd.to_numeric, errors='coerce')
    print(f'Total Respondents: {df[col].dropna().count()}')
    print(f'Respondents Familiar with Show: {ra.dropna().count()}')
    
    percent_familiar = (ra.dropna().count()/df[col].dropna().count()) * 100   
    print(f'Percent Familiar with Show: {percent_familiar:.1f}%')
    
    if ra.dropna().count() > 0:
        print(f'Median: {int(ra.dropna().median())}')
        print(f'Mean: {ra.dropna().mean():.2f}')
        
    print('  ')
    print('  ')
    print('  ')
    

```
